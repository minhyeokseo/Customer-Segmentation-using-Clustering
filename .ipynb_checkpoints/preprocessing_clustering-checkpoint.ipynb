```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
import numpy as np
from sklearn.metrics import silhouette_score
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# data set
customers=pd.read_csv('./data/customers.csv')
order_items=pd.read_csv('./data/order_items.csv')
orders=pd.read_csv('./data/orders.csv')
payments=pd.read_csv('./data/payments.csv')
products=pd.read_csv('./data/products.csv')
```

```python

df_1=pd.merge(customers,orders,on='customer_id',how='left')

df_2=pd.merge(df_1,payments,on='order_id',how='left')

df_3=pd.merge(df_2,order_items,on='order_id',how='left')

f_df=pd.merge(df_3,products,on='product_id',how='left')

f_df.to_csv('./data/.f_df.csv')

f_df.info()

##필요없는 컬럼제거
f_df=f_df.drop(['payment_sequential','order_id','order_item_id','product_id','seller_id'],axis=1)

## 승인 안난 주문 제거
f_df=f_df[~f_df['order_status'].isin(['unavailable','canceled'])]

f_df.isnull().sum()

f_df

#f_df=f_df.dropna(subset=['payment_type','price'])
# 어차피 널값비중 적으니 그냥 다 삭제
f_df=f_df.dropna()

# 중복제거
f_df=f_df.drop_duplicates()



#f_df[round(f_df['price']+f_df['shipping_charges'],2)==f_df['payment_value']][['order_approved_at','price','shipping_charges','payment_value']][:]

#f_df[(f_df['price']+f_df['shipping_charges']!=f_df['payment_value'])][['price','shipping_charges','payment_value']]



#결제금액이 더 큰경우 없애기
f_df=f_df[~(round(f_df['price']+f_df['shipping_charges'],2)<f_df['payment_value'])]

f_df
#세일 변수
f_df['sale']=round(f_df['price']+f_df['shipping_charges'],2)-f_df['payment_value']
#세일 퍼센트 변수
f_df['sale_p']=round(f_df['sale']/(f_df['price']+f_df['shipping_charges'])*100)

f_df

```

```python
# 시계열자료 데이트타임으로변경
f_df['order_purchase_timestamp']=pd.to_datetime(f_df['order_purchase_timestamp'])
f_df['order_approved_at']=pd.to_datetime(f_df['order_approved_at'])
f_df['order_delivered_timestamp']=pd.to_datetime(f_df['order_delivered_timestamp'])
```

```python
#배송차이
f_df['d_day']=(f_df['order_delivered_timestamp']-f_df['order_approved_at']).dt.days
```

```python
#소비자의 주문량
a=f_df.groupby('customer_id').size().reset_index(name='order_count')

#마지막으로 결제한날과의 차이
b=f_df.groupby('customer_id')['order_approved_at'].max().reset_index()
b['recent']=(f_df['order_approved_at'].max()-b['order_approved_at']).dt.days
b=b[['customer_id','recent']]


#지금까지사용한금액
c=f_df.groupby('customer_id')['payment_value'].sum().reset_index()
c
```

```python
a_b=pd.merge(a,b,on='customer_id',how='left')
a_b_c=pd.merge(a_b,c,on='customer_id',how='left')
```

```python
a_b_c
```

```python
#지역 바꾸기
def state_f(x):
    if x in ('AC','RO','AM','RR','AP','PA','TO'):
        return 'North'
    elif x in ('MA','PI','CE','RN','PB','PE','AL','SE','BA'):
        return 'Northeast'
    elif x in ('GO','MT','MS'):
        return 'Midwest'
    elif x in ('MG','ES','RJ','SP'):
        return 'Southeas'
    elif x in  ('PR','SC','RS'):
        return 'South'
    else:
        return 'Federal'
```

```python
f_df['District']=f_df['customer_state'].apply(state_f)

d=f_df.groupby('District')['d_day'].mean().reset_index()

z=f_df[['customer_id','District']]
z=z.drop_duplicates()

tem=pd.merge(a_b_c,z,on='customer_id',how='left')
tem=pd.merge(tem,d,on='District',how='left')
h=f_df.groupby('customer_id')['payment_installments'].mean().reset_index()
tem=pd.merge(tem,h,on='customer_id',how='left')
```

# 왜도가 양수이기때문에 정규분포로 만들어주기 위해  로그변환

```python
for feature in temp.columns[1:2]:
    temp[feature] = temp[feature].apply(lambda x: np.power(x, (1/10)))  #10th root
for feature in temp.columns[3:4]:
    temp[feature] = temp[feature].apply(lambda x: np.power(x, (1/10)))  #10th root
for feature in temp.columns[5:6]:
    temp[feature] = temp[feature].apply(lambda x: np.power(x, (1/10)))  #10th root
for feature in temp.columns[7:]:
    temp[feature] = temp[feature].apply(lambda x: np.power(x, (1/10)))  #10th root
```

```python
temp=temp.drop(['customer_id','sale','District','approval_time'],axis=1)
#temp=temp.drop(['Cluster'],axis=1)
```

```python
temp
```

```python
#col=temp.select_dtypes(exclude='object').copy().columns
col=[ 'order_count','recent', 'payment_value', 'payment_installments']
scaler = StandardScaler()
temp[col] = scaler.fit_transform(temp[col])
# from sklearn.preprocessing import LabelEncoder
# le=LabelEncoder()
# temp['District']=le.fit_transform(temp['District'])
# temp['payment_type']=le.fit_transform(temp['payment_type'])

temp=pd.get_dummies(temp)
```

```python
temp['delivery_delay'].value_counts()
```

```python
temp
```

```python
from sklearn.decomposition import PCA

# 주성분 개수를 판단하기 위한 pca임의 시행 
pca = PCA(n_components=3)
pca.fit(temp)
# 설정한 주성분의 갯수로 전체 데이터 분산을 얼만큼 설명 가능한지 
print(pca.explained_variance_ratio_.sum())

# pca 시행
pca_df = pca.fit_transform(temp)
pca_df = pd.DataFrame(data = pca_df, columns = ['PC1','PC2','PC3'])  

# Show the first 5 firms
pca_df.head()
```

```python
# 초기 k 값 참고를 위한 scree plot 을 그리고, 군집이 나뉘는 시간까지 고려한 k 값 확인 
model = KMeans()

# k 값의 범위를 조정해 줄 수 있습니다. 
visualizer = KElbowVisualizer(model, k=(3,12))

# 데이터 적용 
visualizer.fit(pca_df) 
visualizer.show()  
```

```python
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(pca_df)
    inertia.append(kmeans.inertia_)

plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

```python
#  KMEANS
# 군집개수(n_cluster)는 5,초기 중심 설정방식 랜덤,  
kmeans = KMeans(n_clusters=4, random_state=42,init='random')

# pca df 를 이용한 kmeans 알고리즘 적용
kmeans.fit(pca_df)

# 클러스터 번호 가져오기 
labels = kmeans.labels_

# 클러스터 번호가 할당된 데이터셋 생성
kmeans_df = pd.concat([pca_df, pd.DataFrame({'Cluster':labels})],axis = 1)
```

```python
# 3D 산점도 그리기
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# 각 클러스터 별로 색상을 다르게 표시
scatter = ax.scatter(kmeans_df['PC1'], kmeans_df['PC2'], kmeans_df['PC3'], c=kmeans_df['Cluster'], cmap='Set1', s=100)

ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Feature 3')
ax.set_title('Clustering Visualization in 3D without PCA')

# 범례 추가
legend1 = ax.legend(*scatter.legend_elements(), title='Cluster')
ax.add_artist(legend1)

plt.show()
```

```python
# 2차원으로 시각화

plt.figure(figsize=(8,6))
    
sns.scatterplot(data = kmeans_df, x = 'PC1', y='PC2', hue='Cluster')
plt.title('The Plot Of The Clusters(2D)')
plt.show()
```
